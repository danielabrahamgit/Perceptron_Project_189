### Activation Function
A Perceptron has 3 basic building blocks; an array of inputs from our data, weights for these inputs, and an activation function which gives us our output.

We will break all of this down shortly, but first we will take a look at the activation function.  An activation function is a non-lineaer function that maps a value to a binary output.  There are a lot of different activation functions that can be used, but the one we use for Perceptron is very simple:

\[\begin f(x) = 
        \begin{cases}
        1, x > 0 \\
        0, x \leq 0
        \end{cases} \]
\\
 

Based on the function above, it should be clear that Perceptron is a binary classifier; it labels its data as 0 or 1.  We have started the activation funcion below, fill in the code so that the activation is performed correctly

```python
def activation(x):
    """
    Write the code for the Perceptron activation function.
    Map x to its correct output based on the rules defined above.  Do not overthink.
    
    """
    result = None

    ##Your Code Here##
    result = x>0
    result = 1*result
    
    return result
```

### Inputs and Weights
Now that we have our Activation Function defined, the next question we must ask ourselves is what is this value x that we are passing it?  This value is going to be used to classify our data, so it will need to be a combination of all the information we have recieved as input.  However, not every piece of information is going to be as useful as the others, and this is where the weights come in.  By multiplying each input by its corresponding weight and adding everything up, we get a resulting value that ends up being very useful for classification.  This calculation should sound familiar to you, because it is really just a dot product between the weights and the inputs. 

```python
def summation(inputs, weigths):
    """
    Write the code for calculating the weighted sum that we will pass into the activation function.
    We want to multiply each input with its respective weight, i.e. take the dot product between input and weight.
    
    """
    result = None
    
    ##Your Code Here##
    result = inputs@weights
    
    return result

```

You likely saw a lot of incorrect results from that.  But this is what we expect to see, we chose random weights so of course the Perceptron algorithm isn't going to seperate the data correctly.  So how do we fix this? How do we change our weights so that we get our desired output?  We do this using the Perceptron update formula.  We want to update our weight whenever there is an incorrect classification, so we know that (Y_true - Y_pred) should be in the formula, as this will give us zero if we predict the label correctly.  We also know that every data point should be able to affect this update, because we want even a single missclassification to be able to alter the weights.  This means we will have to loop through all of the data points as wel update.

More specifically, we have the equation:

\begin{equation}
{w}_{new} = {w}_{old} + r * (y_i - y^*_i) * {x}_i
\end{equation}



The terms in the above equation should mostly make sense to you, but at this point you should have 2 questions.  What is r? And when do we stop updating?

r is simply our learning rate, it determines how quickly we are updating our weights.  This is what is know as a hyperparameter, and while you don't need to understand what that means right now, just understand it is a value that can be changed to improve our performace.  We will not be tuning this hyperparameter at this point in the course, we will just set r = .5 and be happy with that.

And as for when to stop updating, this can be done when we have correctly classified all of our data.  We know that our data can be seperated, so we should be fine to let the algorith run to completion.  But in practice we may not know that our data can be seperated, so we would likely want to put some stopping condition into our update equation.  Some common examples of this would be breaking out of the loop after a certain period of time or after our error reaches a sufficiently low value.  Implement the update equation below, and add a stopping measure (will also be helpful for if you write the program incorrectly).


```python
def updateWeights(inputs, weights, correct):
    r = 2
    curr = activation(summation(restaurants, weights))
    count = 0
    while not np.array_equal(curr, correct):
        count += 1
        for x in range(len(restaurants)):
            weights = weights + .5 * (correct[x]-curr[x])*restaurants[x]
        curr = activation(summation(restaurants, weights))
        if count % 100 == 0:
            print(curr)
    return (curr,weights)
    
    
```
